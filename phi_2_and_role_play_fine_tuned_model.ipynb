```python
from transformers import AutoModelForCausalLM
from peft import LoraConfig,PeftModel,AutoPeftModelForCausalLM
#set the LoRA configurations
peft_config = LoraConfig(
        lora_alpha=16,
        lora_dropout=0.05,
        r=16,
        bias="none",
        task_type="CAUSAL_LM",
)
#
peft_model_id = "Laksh99/phi-2-role-play"
config = peft_config.from_pretrained(peft_model_id)
#
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path,
                                             return_dict=True,
                                             load_in_4bit=True,
                                             device_map="auto",
                                             )
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(peft_model_id)
```

```python
tokenizer= AutoTokenizer.from_pretrained(peft_model_id)
#
model = PeftModel.from_pretrained(model,peft_model_id)
#
print(model.get_memory_footprint())
```

```python


from transformers import pipeline


pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    trust_remote_code=True,
)
```

```python
#import logging
#logging.set_verbosity(logging.CRITICAL)

prompt = '''<|system|>Wonder Woman is a warrior princess of the Amazons with a strong sense of justice and a mission.
<|user|> What motivates you to fight for peace and love?
<|assistant|>'''
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=200)
result = pipe(prompt)
print(result[0]['generated_text'])
```

```python
#import logging
#logging.set_verbosity(logging.CRITICAL)

prompt = '''<|system|>You are Sherlock Holmes the famous detective from Baker Street.
<|user|> Give me a brief plot based on a case you solved?
<|assistant|>'''
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=200)
result = pipe(prompt)
print(result[0]['generated_text'])
```
