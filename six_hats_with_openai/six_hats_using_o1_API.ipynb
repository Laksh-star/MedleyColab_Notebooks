```python
from openai import OpenAI

def main():
    # Initialize the OpenAI client (replace with your actual API key if needed)
    client = OpenAI(api_key="your_api_key_here")

    # Send a simple prompt to the chat completion endpoint
    completion = client.chat.completions.create(
        model="o1-mini",   # or "gpt-4", "gpt-3.5-turbo", etc.
        store=True,
        messages=[
            {"role": "user", "content": "write a haiku about ai"}
        ]
    )

    # Extract the response text
    result = completion.choices[0].message.content

    # Print it to screen (optional)
    print("=== Chat Completion ===")
    print(result)

    # Write the result to a Markdown file
    with open("recommendation.md", "w", encoding="utf-8") as f:
        f.write(result)

if __name__ == "__main__":
    main()
```
