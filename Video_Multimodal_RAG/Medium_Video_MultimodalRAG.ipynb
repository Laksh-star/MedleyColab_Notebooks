```python
import os
import cv2
from pytubefix import YouTube
from moviepy.editor import VideoFileClip
import whisper
import lancedb
from langchain_openai import OpenAIEmbeddings
from anthropic import Anthropic
from IPython.display import display, Image
import requests
from urllib.parse import urlparse, parse_qs
from tqdm import tqdm  # Progress bar

# Replace these with your actual API keys
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY", "your-key-here")
ANTHROPIC_API_KEY = os.getenv("ANTHROPIC_API_KEY", "your-key-here")
LANCEDB_HOST_FILE = "./data/.lancedb"

class VideoProcessor:
    def __init__(self):
        self.whisper_model = whisper.load_model("base")

    def download_video(self, youtube_url, output_path, retries=3):
        """
        Downloads the video from YouTube with retries and shows a progress bar.
        """
        print("Downloading video...")
        attempt = 0
        while attempt < retries:
            try:
                yt = YouTube(youtube_url)
                stream = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
                if stream is None:
                    raise Exception("No suitable stream found for the video.")

                # Show progress bar while downloading
                with tqdm(total=100, desc="Downloading video", unit="chunk") as pbar:
                    downloaded_file = stream.download(output_path)
                    pbar.update(100)
                return downloaded_file
            except Exception as e:
                attempt += 1
                print(f"Error downloading video (attempt {attempt}/{retries}): {e}")
                if attempt >= retries:
                    return None
        return None

    def extract_frames(self, video_path, output_path, fps=1):
        """
        Extracts frames from the video at the specified frame rate and shows progress.
        """
        video = cv2.VideoCapture(video_path)
        total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        frame_count = 0
        frames = []

        print(f"Extracting frames (Total frames: {total_frames})...")
        with tqdm(total=total_frames, desc="Extracting frames", unit="frame") as pbar:
            while True:
                success, frame = video.read()
                if not success:
                    break
                if frame_count % int(video.get(cv2.CAP_PROP_FPS) / fps) == 0:
                    frame_path = os.path.join(output_path, f"frame_{frame_count}.jpg")
                    cv2.imwrite(frame_path, frame)
                    frames.append({"path": frame_path, "timestamp": frame_count / video.get(cv2.CAP_PROP_FPS)})
                frame_count += 1
                pbar.update(1)
        video.release()
        return frames

    def extract_audio(self, video_path, output_path):
        video = VideoFileClip(video_path)
        audio_path = os.path.join(output_path, "audio.mp3")
        video.audio.write_audiofile(audio_path)
        return audio_path

    def transcribe_audio(self, audio_path, output_path):
        """
        Transcribes the audio using the Whisper model and saves the transcription to a file.
        Also shows progress during transcription.
        """
        print("Transcribing audio...")

        result = self.whisper_model.transcribe(audio_path)

        # Save transcription to a text file
        transcript_file = os.path.join(output_path, "transcription.txt")
        with open(transcript_file, 'w', encoding='utf-8') as f:
            f.write(result["text"])

        print(f"Transcription saved at: {transcript_file}")
        return result["text"]

    def process(self, youtube_url):
        temp_dir = "./temp"
        os.makedirs(temp_dir, exist_ok=True)

        try:
            # Progress bar for video downloading
            video_path = self.download_video(youtube_url, temp_dir)
            if not video_path:
                raise Exception("Failed to download video after retries.")

            # Progress bar for frame extraction
            frames = self.extract_frames(video_path, temp_dir)

            # Extract audio
            audio_path = self.extract_audio(video_path, temp_dir)

            # Show progress bar while transcribing
            with tqdm(total=100, desc="Transcribing audio") as pbar:
                transcript = self.transcribe_audio(audio_path, temp_dir)
                pbar.update(100)

            return {
                "frames": frames,
                "transcript": transcript,
                "video_path": video_path
            }
        except Exception as e:
            print(f"Error processing video: {e}")
            return None

class VectorStore:
    def __init__(self, uri, openai_api_key):
        self.db = lancedb.connect(uri)
        self.embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)
        self.table_name = "video_data"

    def add_video_data(self, video_data):
        frames = video_data["frames"]
        transcript = video_data["transcript"]

        data = []
        for frame in frames:
            text = f"Frame at {frame['timestamp']} seconds: {transcript}"
            vector = self.embeddings.embed_query(text)
            data.append({
                "vector": vector,
                "text": text,
                "image_path": frame["path"],
                "metadata": {
                    "timestamp": frame["timestamp"],
                    "video_path": video_data["video_path"]
                }
            })

        table = self.db.create_table(self.table_name, data=data, mode="overwrite")

    def search(self, query, k=3):
        table = self.db.open_table(self.table_name)
        query_vector = self.embeddings.embed_query(query)
        results = table.search(query_vector).limit(k).to_list()
        return results

class MultimodalRetriever:
    def __init__(self, vector_store):
        self.vector_store = vector_store

    def retrieve(self, query, k=3):
        return self.vector_store.search(query, k)

class LVLMInference:
    def __init__(self, api_key):
        self.client = Anthropic(api_key=api_key)

    def generate(self, prompt, image_path=None):
        messages = [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt
                    }
                ]
            }
        ]

        if image_path:
            with open(image_path, "rb") as img:
                img_data = img.read()
                messages[0]["content"].append({
                    "type": "image",
                    "image_data": {
                        "mime_type": "image/jpeg",
                        "data": img_data
                    }
                })

        response = self.client.messages.create(
            model="claude-3-sonnet-20240229",
            max_tokens=1000,
            messages=messages
        )

        return response.content[0].text

class MultimodalRAGPipeline:
    def __init__(self, retriever, lvlm):
        self.retriever = retriever
        self.lvlm = lvlm

    def run(self, query):
        retrieved_data = self.retriever.retrieve(query)

        context = "\n".join([item["text"] for item in retrieved_data])
        prompt = f"Based on the following context, please answer the question: {query}\n\nContext: {context}"

        answer = self.lvlm.generate(prompt)

        return {
            "answer": answer,
            "retrieved_data": retrieved_data
        }

class MultimodalRAGSystem:
    def __init__(self):
        self.video_processor = VideoProcessor()
        self.vector_store = None
        self.retriever = None
        self.lvlm = LVLMInference(ANTHROPIC_API_KEY)
        self.pipeline = None
        self.video_processed = False

    def process_video(self, youtube_url):
        video_data = self.video_processor.process(youtube_url)
        if video_data:
            try:
                self.vector_store = VectorStore(LANCEDB_HOST_FILE, OPENAI_API_KEY)
                self.vector_store.add_video_data(video_data)
                self.retriever = MultimodalRetriever(self.vector_store)
                self.pipeline = MultimodalRAGPipeline(self.retriever, self.lvlm)
                self.video_processed = True
                print("Video processed successfully!")
            except Exception as e:
                print(f"Error creating vector store: {e}")
                self.video_processed = False
        else:
            print("Failed to process the video.")
            self.video_processed = False

    def answer_question(self, question):
        if not self.video_processed:
            print("No video has been successfully processed. Please process a video first.")
            return

        try:
            result = self.pipeline.run(question)
            print("Answer:", result["answer"])
            print("\nRetrieved Frames:")
            for item in result["retrieved_data"]:
                display(Image(filename=item["image_path"]))
                print(f"Timestamp: {item['metadata']['timestamp']} seconds")
                print(item["text"])
                print("---")
        except Exception as e:
            print(f"Error answering question: {e}")

# Interactive part for terminal-based interaction
system = MultimodalRAGSystem()

# Process video loop
while True:
    youtube_url = input("Enter YouTube URL (or type 'quit' to exit): ")
    if youtube_url.lower() == 'quit':
        print("Exiting video processing.")
        break
    system.process_video(youtube_url)
    if system.video_processed:
        break

# Answer questions loop
while True:
    question = input("Ask a question about the video (or type 'quit' to exit): ")
    if question.lower() == 'quit':
        print("Exiting question-answer session.")
        break
    system.answer_question(question)
```
